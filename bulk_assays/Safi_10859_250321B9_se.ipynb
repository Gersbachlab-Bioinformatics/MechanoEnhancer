```python
%%writefile /hpc/group/gersbachlab/aeb84/reddylab_utils/scripts/ataqc.make_tss_plot.py
#!/usr/bin/env python

from os import path
import argparse

def main():
    parser = argparse.ArgumentParser()
    parser.add_argument('--tss', help="TSS reference file")
    parser.add_argument('--bam', help="BAM file")
    parser.add_argument('--read-length', type=int, default=50, help="Read length")
    parser.add_argument('--processes', type=int, default=1, help="Number of threads (cpus in SLURM) to use.")
    parser.add_argument('--outfile-prefix', help="Rootname to output TSS plot")
    parser.add_argument('--chromosome-sizes', help="Chromosome sizes file")
    parser.add_argument('--ataqc-rootpath', help="Root path to ATAQC package", default="/data/reddylab/software")

    args = parser.parse_args()

    # Load ATAQC functions
    path.sys.path.append(args.ataqc_rootpath)
    from ataqc import run_ataqc

    plotfile, plotfile_large, tss_enrich_point_val = \
        run_ataqc.make_tss_plot(bam_file=args.bam,
                                tss=args.tss,
                                prefix=args.outfile_prefix,
                                chromsizes=args.chromosome_sizes,
                                read_len=args.read_length,
                                processes=args.processes)

    print("Done!")
    print("Save files:")
    print("\t%s\n\t%s" % (plotfile, plotfile_large))
    print("%s TSS enrichemnt point value:" % args.bam, tss_enrich_point_val)

if __name__ == "__main__":
    main()
```

```python
%%writefile /hpc/group/gersbachlab/aeb84/software/cwl/generate_stats_atacseq_single_end.py
#!/usr/bin/env python
from collections import defaultdict as dd
import argparse

parser = argparse.ArgumentParser(description='Generate QC stats for the ChIP-seq CWL pipeline')
parser.add_argument('outdir', type=str, help='Directory with CWL ChIP-seq pipeline output files')
parser.add_argument('-samples', required=True, nargs='+', help='Sample names for which stats are computed')
parser.add_argument('--nonmasked', required=False, action='store_true', help='If no blacklist used after mapping')

args = parser.parse_args()
FILES_DIR=args.outdir
samples=args.samples

stats_files_ext = {
    'raw'    : 'read_count.txt',
    'trimmed': 'trimmed.read_count.txt',
    'mapped' : 'trimmed.bowtie.log.read_count.mapped',
    'mapped_and_filtered': 'trimmed.masked.dedup.sorted_read_count.txt',
    'in_broadpeaks'  : 'NA',
    'in_narrowpeaks'  : 'trimmed.masked.dedup.sorted.bam.in_peaks.read_count.within_replicate.txt',
    'broadPeak_count': 'trimmed.masked.dedup.sorted_peaks.broadPeak.peak_count.within_replicate.txt',
    'narrowPeak_count': 'trimmed.masked.dedup.sorted_peaks.narrowPeak.peak_count.within_replicate.txt',
    'NRF': 'trimmed.preseq_c_curve.percentage_unique_reads.txt',
    'PBC': 'trimmed.PBC.txt',
    'NSC': 'trimmed.masked.dedup.sorted.spp_cross_corr.txt'
}

if args.nonmasked:
    stats_files_ext = {k: v.replace('.masked', '') for k,v in stats_files_ext.items()}

stats_sorted = [
    'raw',
    'trimmed',
    'mapped',
    'percentage_unique',
    'mapped_and_filtered',
    'percentage_unique_mapped_and_filtered',
    'in_broadpeaks',
    'in_narrowpeaks',
    'percent_in_broadpeaks',
    'percent_in_narrowpeaks',
    'broadPeak_count',
    'narrowPeak_count',
    'NRF',
    'PBC',
    'NSC',
    'RSC'
]

print(','.join(['sample'] + stats_sorted))

for s in samples:
    stats=dd(lambda: 'N/A')
    for k in stats_sorted:  # stats_files_ext.keys():
        try:
            with open('%s/%s.%s'%(FILES_DIR, s, stats_files_ext[k])) as ifile:
                for l in ifile:
                    vals = l.rstrip().split()
                    if k == 'mapped':
                        stats[k] = vals[1]
                        try:
                            stats['percentage_unique'] = 100*float(vals[1])/float(vals[0])
                        except TypeError: 
                            stats['percentage_unique'] = 'N/A'
                    elif k == 'NSC':
                        stats['NSC']=float(vals[8])
                        stats['RSC']=float(vals[9])
                    elif k == 'in_narrowpeaks' or k == 'broadpeaks': 
                        stats[k] = int(vals[0])
                        try:
                            stats['percent_%s' % k]=100*float(stats[k])/float(stats['mapped_and_filtered'])
                        except TypeError as ValueError:
                            stats['percent_%s' % k]='N/A'
                    else:
                        try:
                            stats[k]=int(vals[0])
                        except:
                            stats[k]=float(vals[0])
        except (IOError, KeyError, ValueError):
            pass
    try:
        stats['percentage_unique_mapped_and_filtered'] = 100*float(stats['mapped_and_filtered'])/stats['trimmed']
    except ValueError:
        stats['percentage_unique_mapped_and_filtered'] = 'N/A'
    print(','.join([s] + [str(stats[kk]) for kk in stats_sorted]))
```
